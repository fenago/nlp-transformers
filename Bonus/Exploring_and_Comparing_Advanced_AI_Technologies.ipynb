## 2.1.Google AI Cloud Language API

https://cloud.google.com/natural-language/docs/basics#sentiment_analysis


````python
from google.cloud import language

def analyze_text_sentiment(text: str):
    client = language.LanguageServiceClient()
    document = language.Document(content=text, type_=language.Document.Type.PLAIN_TEXT)

    response = client.analyze_sentiment(document=document)

    sentiment = response.document_sentiment
    results = dict(
        text=text,
        score=f"{sentiment.score:.1%}",
        magnitude=f"{sentiment.magnitude:.1%}",
    )
    for key, value in results.items():
        print(f"{key:10}: {value}")
````

````python
# Easy
text = "The Super Mario Bros. Movie was a great picture!"
analyze_text_sentiment(text)
````

````python
# Difficult : the movie is great but the analysis is neutral
text = "The Super Mario Bros. Movie was a great picture, however I expected more action and did not like some scenes although overall it was still great."
analyze_text_sentiment(text)
````

## 2.2.OpenAI GPT-4 API

````python
 def sentiment_analysis(uinput):
   #preparing the prompt 
   role="user"
   line = {"role": role, "content": uinput}
   #creating the message
   assert1={"role": "system", "content": "You are a sentiment analysis agent."}
   assert2={"role": "assistant", "content": "You provide a sentiment analysis score and a sentiment analysis magnitude."}
   assert3=line
   iprompt = []
   iprompt.append(assert1)
   iprompt.append(assert2)
   iprompt.append(assert3)

   response=openai.ChatCompletion.create(model="gpt-4",messages=iprompt) #GPT-4 dialog
   text=response["choices"][0]["message"]["content"] #response in JSON
   return text
````

# 3.Entity analysis



## 3.1. Google AI Cloud Language API

https://cloud.google.com/natural-language/docs/basics#entity_analysis

````python
from google.cloud import language


def analyze_text_entities(text: str):
    client = language.LanguageServiceClient()
    document = language.Document(content=text, type_=language.Document.Type.PLAIN_TEXT)

    response = client.analyze_entities(document=document)

    for entity in response.entities:
        print("=" * 80)
        results = dict(
            name=entity.name,
            type=entity.type_.name,
            salience=f"{entity.salience:.1%}",
            wikipedia_url=entity.metadata.get("wikipedia_url", "-"),
            mid=entity.metadata.get("mid", "-"),
        )
        for key, value in results.items():
            print(f"{key:15}: {value}")
            
````

````python
# movie found as of May 1, 2023
text = "The Super Mario Bros. Movie was a great picture, however I expected more action and did not like some scenes although overall it was still great!"
analyze_text_entities(text)
````

## 4.1.Google AI Cloud Language API


https://cloud.google.com/natural-language/docs/basics#syntactic_analysis

````python
from google.cloud import language


def analyze_text_syntax(text: str):
    client = language.LanguageServiceClient()
    document = language.Document(content=text, type_=language.Document.Type.PLAIN_TEXT)

    response = client.analyze_syntax(document=document)

    line = "{:10}: {}"
    print(line.format("sentences", len(response.sentences)))
    print(line.format("tokens", len(response.tokens)))
    for token in response.tokens:
        print(line.format(token.part_of_speech.tag.name, token.text.content))
````

````python
text = "The Super Mario Bros. Movie was a great picture, however I expected more action and did not like some scenes although overall it was still great!"
analyze_text_syntax(text)
````

## 5.1.Google AI Cloud Language API

````python
from google.cloud import language


def classify_text(text: str):
    client = language.LanguageServiceClient()
    document = language.Document(content=text, type_=language.Document.Type.PLAIN_TEXT)

    response = client.classify_text(document=document)

    for category in response.categories:
        print("=" * 80)
        print(f"category  : {category.name}")
        print(f"confidence: {category.confidence:.0%}")
        
````

````python
text = ("The Super Mario Bros. Movie was a great picture, however I expected more action and did not like some scenes although overall it was still great!")
classify_text(text)
````

````python
import gradio as gr
from transformers import pipeline

# Load the sentiment analysis model
sentiment_pipeline = pipeline("sentiment-analysis")

# Define the function to be called when the user submits input
def analyze_sentiment(text):
    result = sentiment_pipeline(text)
    return result[0]["label"]

# Create the Gradio interface
iface = gr.Interface(
    fn=analyze_sentiment,
    inputs=gr.inputs.Textbox(lines=5, placeholder="Enter text here..."),
    outputs="text",
    title="Sentiment Analysis",
    description="Enter a text and the model will predict its sentiment (positive or negative).",
)

# Launch the interface
iface.launch()
````

### 7.1.2. OpenAI

````python
import gradio as gr
import openai

# Define a function to perform sentiment analysis
def analyze_sentiment(text):
    response = openai.Completion.create(
        engine="gpt-3.5-turbo-instruct",
        prompt=f"Sentiment analysis of the following text: '{text}'\n",
        max_tokens=100,
        n=1,
        stop=None,
        temperature=0.5,
    )
    sentiment = response.choices[0].text.strip()
    return sentiment

# Create the Gradio interface
iface = gr.Interface(
    fn=analyze_sentiment,
    inputs=gr.inputs.Textbox(lines=5, label="Enter text here..."),
    outputs=gr.outputs.Textbox(label="Sentiment analysis"),
    title="Sentiment Analysis",
    description="Enter a text and the model will predict the sentiment of the text.",
)

# Launch the interface
iface.launch()
````

## 7.2. Entity Recognition

### 7.2.1. Hugging Face

````python
import gradio as gr
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

# Create the named entity recognition pipeline
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)

# Define a function to analyze entities
def analyze_entities(text):
    entities = ner_pipeline(text)
    formatted_entities = []
    for entity in entities:
        word = entity['word']
        if word.startswith("##"):
            word = word[2:]
        formatted_entities.append((word, entity['entity']))
    result = " ".join([f"{word} ({entity_type})" for word, entity_type in formatted_entities])
    return result

# Create the Gradio interface
iface = gr.Interface(
    fn=analyze_entities,
    inputs=gr.inputs.Textbox(lines=5, placeholder="Enter text here..."),
    outputs="text",
    title="Named Entity Recognition",
    description="Enter a text and the model will identify named entities and their types (e.g., person, organization, location).",
)

# Launch the interface
iface.launch()
````
